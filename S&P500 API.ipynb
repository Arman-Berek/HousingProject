```python
import requests
import json
import pandas as pd
```

```python
url = "https://api.marketdata.app/v1/indices/candles/D/SPX/?from=1975-12-31&to=2023-01-01&dateformat=timestamp"
```

```python
headers1 = {'Authorization':'Token WU1XRFptV1hMdXdjSUxNOUpzaWdQMjhoWmNZZlg1V2d3UG1XMko3ZHdQND0'}
```

```python
response = requests.request("GET", url, headers = headers1)
```

```python
response.text
```

```python
json_text = json.loads(response.text)
```

```python
json_text.keys()
```

```python
spx_df = pd.DataFrame(json_text)
```

```python
spx_df2 = spx_df.drop('s', axis = 1)
```

```python
#renaming columns:
columns = {'o':'Open',
           'c':'Close',
           'h':'High',
           'l':'Low',
           'v':'Volume',
           't':'Time'
          }

spx_final = spx_df2.rename(columns = columns)
```

```python
spx_final.head()
```

```python
#dataframe to CSV
spx_final.to_csv('S&P500.csv', index = False)
```
